```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.columns
```

```python
final =pd.merge(df,df_country, on = 'Country Code', how='left')
```

```python
final.head()
```

```python
country_names=final.Country.value_counts().index
```

```python
country_names
```

```python
country_val = final.Country.value_counts().values
```

```python
# pie chart
plt.pie(country_val[:2], labels = country_names[:2],autopct ='%1.2f%%' )
```

zomato maximumm records and transaction are from India shown using pie chart

```python
final.columns
```

```python
ratings = final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating Count", data = ratings)
```

```python
import matplotlib                               
matplotlib.rcParams['figure.figsize'] = (15,8)                               # to increase size
sns.barplot(x = "Aggregate rating", y = "Rating Count", data = ratings)
```

```python
import matplotlib                               
matplotlib.rcParams['figure.figsize'] = (15,8)                               # to increase size
sns.barplot(x = "Aggregate rating", y = "Rating Count", data = ratings, palette = ['White', 'red', 'orange', 'yellow','green' ])
```

```python
sns.countplot(x = "Rating color", data= ratings, palette = ['White', 'red', 'orange', 'yellow','green' ])
```
